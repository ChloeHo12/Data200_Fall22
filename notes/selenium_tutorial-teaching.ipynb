```python
# This scrapes one page with ideas above
from selenium import webdriver
import pandas as pd
import time 

driver = webdriver.Chrome('/users/bilene/downloads/chromedriver.exe')
driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/p1')

# get comment ids as a list to iterate on
ids = driver.find_elements("xpath","//*[contains(@id,'Comment_')]")
comment_ids = []
for i in ids:
    comment_ids.append(i.get_attribute('id'))

comments=[]
for x in comment_ids:
    print(x) # to see iterations
    #Extract user ids from each user on a page
    userid_element = driver.find_element('xpath','//*[@id="' + x + '"]/div/div[2]/div[1]/span[1]/a[2]')
    userid = userid_element.text
    
    #Extract dates from for each user on a page
    user_date = driver.find_element('xpath','//*[@id="' + x + '"]/div/div[2]/div[2]/span/a/time')
    date = user_date.get_attribute('title')

    #Extract Message for each user on a page
    user_message = driver.find_element('xpath','//*[@id="' + x +'"]/div/div[3]/div/div[1]')
    comment = user_message.text
                                   
    #Adding date, userid and comment for each user in a dataframe    
    comments.append([date,userid,comment])
    
df_single = pd.DataFrame(comments,columns=['date','userid','comment'])
```

```python
# This does a multi-pager.
from selenium import webdriver
import pandas as pd
import time 
import random

driver = webdriver.Chrome(r'C:\users\eren\desktop\chromedriver.exe')

ls = ['p1','p2']
comments=[]
for j in ls:
    driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/p1')

    # get comment ids as a list to iterate on
    ids = driver.find_elements("xpath","//*[contains(@id,'Comment_')]")
    comment_ids = []
    for i in ids:
        comment_ids.append(i.get_attribute('id'))

    comments=[]
    for x in comment_ids:
        print(x, j) # to see iterations
        #Extract user ids from each user on a page
        userid_element = driver.find_element('xpath','//*[@id="' + x + '"]/div/div[2]/div[1]/span[1]/a[2]')
        userid = userid_element.text

        #Extract dates from for each user on a page
        user_date = driver.find_element('xpath','//*[@id="' + x + '"]/div/div[2]/div[2]/span/a/time')
        date = user_date.get_attribute('title')

        #Extract Message for each user on a page
        user_message = driver.find_element('xpath','//*[@id="' + x +'"]/div/div[3]/div/div[1]')
        comment = user_message.text

        #Adding date, userid and comment for each user in a dataframe    
        comments.append([date,userid,comment])  
         
       
    #time.sleep(1) or
    time.sleep(random.uniform(0, 1))
    
df_out = pd.DataFrame(comments,columns=['date','userid','comment'])
df_out
```
