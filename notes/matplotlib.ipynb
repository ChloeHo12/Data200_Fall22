```python
import matplotlib.pyplot as plt
import pandas as pd

parks = pd.read_csv('https://raw.githubusercontent.com/ernbilen/Data200_Fall22/main/data/national_parks.csv')
```

```python
parks
```

```python
# Create a histogram
plt.figure(figsize=(7,4))

plt.hist(parks.GrandCanyon,bins=10, edgecolor='black', facecolor='springgreen')
plt.title('Histogram of Grand Canyon Visitor Count')

plt.xlabel('# of visitors')
plt.ylabel('count')

plt.savefig(r'C:\Users\bilene\Downloads\grand-canyon.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# multiple histograms on the same graph
plt.hist(parks.GrandCanyon,bins=10, edgecolor='black', \
                     facecolor='springgreen',label='GrandCanyon',alpha=.7)

plt.hist(parks.BryceCanyon,bins=10, edgecolor='black', \
                     facecolor='dodgerblue',label='BryceCanyon',alpha=.7)

plt.xlabel('# of visitors')
plt.ylabel('count')
plt.legend(loc='upper right')

plt.savefig(r'C:\Users\bilene\Downloads\grand-canyon-bryce-canyon.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# multiple histograms on side-by-side graphs
plt.figure(figsize=(8,3))

plt.subplot(121)
plt.hist(parks.GrandCanyon,bins=10, edgecolor='black', \
                     facecolor='springgreen',label='GrandCanyon')
plt.xlim(0,7000000)
plt.legend(loc='upper right')
plt.xlabel('# of visitors')
plt.ylabel('count')

plt.subplot(122)
plt.hist(parks.BryceCanyon,bins=10, edgecolor='black', \
                     facecolor='dodgerblue',label='BryceCanyon')
plt.xlim(0,7000000)
plt.legend(loc='upper right')
plt.xlabel('# of visitors')


plt.show()
```

```python
# A for loop implementation

varlist = ['GrandCanyon','BryceCanyon']
colorlist= ['springgreen','dodgerblue']

plt.figure(figsize=(8,3))

for i in range(2):
    plt.subplot(1,2,i+1)
    plt.hist(parks[varlist[i]], edgecolor='black', facecolor=colorlist[i],label=varlist[i])
    plt.xlim(0,7000000)
    plt.legend(loc='upper right')
    plt.xlabel('# of visitors')
    if i==0:
        plt.ylabel('count')
    
    
plt.savefig(r'C:\Users\bilene\Downloads\grand-canyon-bryce-canyon-sidebyside.png', dpi=300, bbox_inches='tight')
plt.show()
    
```
